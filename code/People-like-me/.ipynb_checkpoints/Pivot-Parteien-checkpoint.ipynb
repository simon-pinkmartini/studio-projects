```python
import pandas as pd
```

```python
df = pd.read_excel('ModifiedData/px-x-1702020000_105 - Parteien - filledup.xlsx')
```

```python
df.head(2)
```

```python
df_pivot = df.pivot(index='Pivot', columns="Partei", values="Parteistimmen").reset_index()
```

```python
df_pivot.head(2)
```

```python
df_pivot['Nummer'] = df_pivot['Pivot'].str.extract('(.*)\|')
```

```python
df_pivot['Einheit'] = df_pivot['Pivot'].str.extract('\|(.*)')
```

```python
df_pivot.pop("Pivot")
```

```python
df_pivot.head()
```

```python
df_pivot.to_csv('ModifiedData/px-x-0102010000_103 - Parteien - pivoted.csv', index=False)
```
