```python
#Cantons
df_kant = pd.read_csv('StructuredData/kant-Haushaltsgroesse.csv')
df_kant.head(2)
```

## Load the Geodata

```python
#One for switzerland
gdf_CH = gpd.read_file('shp/g1l15.shp')
gdf_CH.head(2)
```

```python
#One for the cantons
gdf_kant = gpd.read_file('shp/g1k15.shp')
gdf_kant.head(2)
```

```python
gdf_kant.shape
```

```python
#One for the districts
gdf_bez = gpd.read_file('shp/g1b15.shp')
gdf_bez.head(2)
```

```python
gdf_bez.shape
```

```python
#This shapefile is for the communes
gdf_gem = gpd.read_file('shp/g1g15.shp')
gdf_gem.head(2)
```

```python
gdf_gem.shape
```

## Merge with Geodata

```python
#Cantons
gdf_kant_data = gdf_kant.merge(df_kant, how='inner', left_on='KTNR', right_on='Kanton_ID')
gdf_kant_data.pop('AREA_HA')
gdf_kant_data.pop('GRNR')
gdf_kant_data.pop('KTNAME')
gdf_kant_data.pop('KTNR')
gdf_kant_data.pop('X_CNTR')
gdf_kant_data.pop('X_MAX')
gdf_kant_data.pop('X_MIN')
gdf_kant_data.pop('Y_CNTR')
gdf_kant_data.pop('Y_MAX')
gdf_kant_data.pop('Y_MIN')
gdf_kant_data.pop('Z_AVG')
gdf_kant_data.pop('Z_MAX')
gdf_kant_data.pop('Z_MED')
gdf_kant_data.pop('Z_MIN')
gdf_kant_data.head(2)
```

```python
#Districts
gdf_bez_data = gdf_bez.merge(df_bez, how='inner', left_on='BZNR', right_on='Bezirk_ID')
gdf_bez_data.pop('AREA_HA')
gdf_bez_data.pop('GRNR')
gdf_bez_data.pop('KTNR')
gdf_bez_data.pop('BZNAME')
gdf_bez_data.pop('BZNR')
gdf_bez_data.pop('X_CNTR')
gdf_bez_data.pop('X_MAX')
gdf_bez_data.pop('X_MIN')
gdf_bez_data.pop('Y_CNTR')
gdf_bez_data.pop('Y_MAX')
gdf_bez_data.pop('Y_MIN')
gdf_bez_data.pop('Z_AVG')
gdf_bez_data.pop('Z_MAX')
gdf_bez_data.pop('Z_MED')
gdf_bez_data.pop('Z_MIN')
gdf_bez_data.head(2)
```

```python
#Communes
gdf_gem_data = gdf_gem.merge(df_gem, how='inner', left_on='GMDNR', right_on='Gemeinde_ID')
gdf_gem_data.pop('AREA_HA')
gdf_gem_data.pop('Shape_Area')
gdf_gem_data.pop('Shape_Leng')
gdf_gem_data.pop('GRNR')
gdf_gem_data.pop('KTNR')
gdf_gem_data.pop('BZNR')
gdf_gem_data.pop('GMDNR')
gdf_gem_data.pop('GMDNAME')
gdf_gem_data.pop('OBJECTID')
gdf_gem_data.pop('X_CNTR')
gdf_gem_data.pop('X_MAX')
gdf_gem_data.pop('X_MIN')
gdf_gem_data.pop('Y_CNTR')
gdf_gem_data.pop('Y_MAX')
gdf_gem_data.pop('Y_MIN')
gdf_gem_data.pop('Z_AVG')
gdf_gem_data.pop('Z_MAX')
gdf_gem_data.pop('Z_MED')
gdf_gem_data.pop('Z_MIN')
gdf_gem_data.pop('Z_CNTR')
gdf_gem_data.head(10)
```

## Test Plots

```python
#Cantons
plt = gdf_kant_data.plot(column='1', scheme='equal_intervals', cmap='PiYG', figsize=(15,10), linewidth=0.25, legend=True)
plt.set_title("Over- / underrepresentation of 1-person households relative to Swiss mean", fontsize=18)
```

```python
#Districts
plt = gdf_bez_data.plot(column='1', scheme='equal_intervals', cmap='PiYG', figsize=(15,10), linewidth=0.25, legend=True)
plt.set_title("Over- / underrepresentation of 1-person households relative to Swiss mean", fontsize=18)
```

```python
#Communes
plt = gdf_gem_data.plot(column='1', scheme='quantiles', cmap='PiYG', figsize=(15,10), linewidth=0.25, legend=True)
plt.set_title("Over- / underrepresentation of 1-person households relative to Swiss mean", fontsize=18)
```
