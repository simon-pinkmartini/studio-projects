### Bla Bla Meter

```python
import requests
```

```python
from bs4 import BeautifulSoup
```

```python
url = 'http://www.blablameter.de/index.php'
```

```python
params = {
    'Text testen': 'Text testen',
    'bc_ip': ''
}
```

```python
def get_blablameter(row):
    params['bc_ip'] = row['text']
    response = requests.post(url, data=params)
    doc = BeautifulSoup(response.text, "html.parser")
    return doc.find('span', class_='info_index2').text
```

```python
df['blablameter'] = df.apply(get_blablameter, axis=1)
```

```python
df['blablameter'] = df['blablameter'].astype(float)
```

### When did the SNB talk about...

```python
def get_wordcounts(words):
    wordcounts = []
    for i in df.index:
        current = {}
        current['date'] = i
        for word in words:
            regex = r'\b' + word + r'\w*\b'
            current[word] = len(re.findall(regex, df.loc[i]['text'], re.IGNORECASE))
        wordcounts.append(current)
    df_wordcounts = pd.DataFrame(wordcounts)
    df_wordcounts = df_wordcounts.set_index('date')
    return df_wordcounts
```

```python
def get_wordcounts_relative(words):
    wordcounts = []
    for i in df.index:
        current = {}
        current['date'] = i
        for word in words:
            regex = r'\b' + word + r'\w*\b'
            wordcount = len(re.findall(regex, df.loc[i]['text'], re.IGNORECASE))
            nowords = df.loc[i]['nowords']
            current[word] = wordcount / nowords            
        wordcounts.append(current)
    df_wordcounts = pd.DataFrame(wordcounts)
    df_wordcounts = df_wordcounts.set_index('date')
    return df_wordcounts
```

```python
def get_wordcount_groups(wordgroups):
    wordcounts = []
    for i in df.index:
        current = {}
        current['date'] = i
        for wordgroup in wordgroups:
            wordcount = 0
            for word in wordgroup:
                #print (word)
                regex = r'\b' + word + r'\w*\b'
                wordcount = wordcount + len(re.findall(regex, df.loc[i]['text'], re.IGNORECASE))
                #wordcount = 0
            current[str(wordgroup)] = wordcount
        wordcounts.append(current)
    df_wordcounts = pd.DataFrame(wordcounts)
    df_wordcounts = df_wordcounts.set_index('date')
    return df_wordcounts
```

```python
def get_wordcount_groups_relative(wordgroups):
    wordcounts = []
    for i in df.index:
        current = {}
        current['date'] = i
        for wordgroup in wordgroups:
            wordcount = 0
            for word in wordgroup:
                #print (word)
                regex = r'\b' + word + r'\w*\b'
                wordcount = wordcount + len(re.findall(regex, df.loc[i]['text'], re.IGNORECASE))
            nowords = df.loc[i]['nowords']
            wordcount = wordcount / nowords
            current[str(wordgroup)] = wordcount
        wordcounts.append(current)
    df_wordcounts = pd.DataFrame(wordcounts)
    df_wordcounts = df_wordcounts.set_index('date')
    return df_wordcounts
```

Change in statement structure over time

```python
df.head(2)
```

```python
def get_semanticsimilarity(column):
    semantic_similarities = []
    for i in range(0, len(df.index)):
        if i != 0:
            current = {}
            current['date'] = df.index[i]           
            #make 3 lists
            now_list = df.iloc[i][column]
            last_list = df.iloc[i-1][column]
            all_list = now_list + last_list
            #dividend
            dividend = 0
            for word in all_list:
                dividend += now_list[word] * last_list[word]
            #divisor
            divisor_now = 0
            for word in all_list:
                divisor_now += now_list[word] ** 2
            divisor_now = divisor_now ** 0.5
            divisor_last = 0
            for word in all_list:
                divisor_last += last_list[word] ** 2
            divisor_last = divisor_last ** 0.5
            divisor = divisor_now * divisor_last
            #quotient
            quotient = dividend / divisor
            current['semanticsimilarity'] = quotient
            semantic_similarities.append(current)
    #cram it into a df
    df_semantic_similarities = pd.DataFrame(semantic_similarities)
    df_semantic_similarities = df_semantic_similarities.set_index('date')
    return df_semantic_similarities
```

```python
def get_semanticsimilarity_yoy(column):
    semantic_similarities = []
    for i in range(0, len(df.index)):
        if i >= 7:
            current = {}
            current['date'] = df.index[i]           
            #make 3 lists
            now_list = df.iloc[i][column] + df.iloc[i-1][column] + df.iloc[i-2][column] + df.iloc[i-3][column]
            last_list = df.iloc[i-4][column] + df.iloc[i-5][column] + df.iloc[i-6][column] + df.iloc[i-7][column]
            all_list = now_list + last_list
            #dividend
            dividend = 0
            for word in all_list:
                dividend += now_list[word] * last_list[word]
            #divisor
            divisor_now = 0
            for word in all_list:
                divisor_now += now_list[word] ** 2
            divisor_now = divisor_now ** 0.5
            divisor_last = 0
            for word in all_list:
                divisor_last += last_list[word] ** 2
            divisor_last = divisor_last ** 0.5
            divisor = divisor_now * divisor_last
            #quotient
            quotient = dividend / divisor
            current['semanticsimilarity'] = quotient
            semantic_similarities.append(current)
    #cram it into a df
    df_semantic_similarities = pd.DataFrame(semantic_similarities)
    df_semantic_similarities = df_semantic_similarities.set_index('date')
    return df_semantic_similarities
```

```python
#this doesnt take into account that statement lenght varied before 2011
df_semantic_similarities = get_semanticsimilarity('wordlist')
ax = df_semantic_similarities.plot(title="Semantic similarity over time", figsize=(15,7), linewidth=4)
ax.set_title("The SNB's statements are becoming more similar", fontsize=20)
ax.set_ylabel("Semantic similarity (QoQ)", fontsize=14)
ax.yaxis.set_ticks_position('none')
ax.set_xlabel("Date of statement", fontsize=14)
ax.xaxis.set_major_locator(dates.YearLocator())
ax.xaxis.set_ticks_position('none')
ax.legend(fontsize=14, borderpad=1, loc=2)
ax.grid(True)
```

```python
df_semantic_similarities_yoy = get_semanticsimilarity_yoy('wordlist')
ax = df_semantic_similarities_yoy.plot(figsize=(15,7), linewidth=4)
ax.set_title("The SNB's statements are becoming more similar", fontsize=30)
ax.set_ylabel("Semantic similarity (4Q YoY)", fontsize=14)
ax.set_xlabel("Date of statement", fontsize=14)
ax.xaxis.set_major_locator(dates.YearLocator())
ax.legend(fontsize=14, borderpad=1, loc=2)
ax.grid(True)
```
