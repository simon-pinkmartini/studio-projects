### Clean the Text

Get a list of all the txt files in the directory

```python
lagebeurteilungen = []
```

Open the files, store in dataframe

Remove letter head, adress info, date of letter

```python
def clean(row):
    testtext = row['cleantext']
    #Titles
    testtext = re.sub(r'\nMedienmitteilung.*\n', '\n', testtext)
    testtext = re.sub(r'\ng\n+n\n+u\n+l\n+i\n+e\n+t\n+t\n+i\n+m\n+n\n+e\n+i\n+d\n+e\n+M', '\n', testtext)
    testtext = re.sub(r'\nKommunikation.*\n', '\n', testtext)
    testtext = re.sub(r'\nPressemitteilung.*\n', '\n', testtext)
    #Adress Stuff
    testtext = re.sub(r'\nPressestelle.*\n', '\n', testtext)
    testtext = re.sub(r'\nPostfach.*\n', '\n', testtext)
    testtext = re.sub(r'\nTelefon.*\n', '\n', testtext)
    testtext = re.sub(r'\nTelefax.*\n', '\n', testtext)
    testtext = re.sub(r'\nFax.*\n', '\n', testtext)
    testtext = re.sub(r'\nwww\.snb\.ch.*\n', '\n', testtext)
    testtext = re.sub(r'\nsnb@snb\.ch.*\n', '\n', testtext)
    testtext = re.sub(r'\ncommunications@snb\.ch.*\n', '\n', testtext)
    #Date of letter
    testtext = re.sub(r'\nZürich,.*\n', '\n', testtext)
    testtext = re.sub(r'\nBern,.*\n', '\n', testtext)
    testtext = re.sub(r'\nGenf,.*\n', '\n', testtext)
    testtext = re.sub(r'\n\d+\.\s\w+\s\d{4}\s*\n', '\n', testtext)
    testtext = re.sub(r'\nabcdefg.*\n', '\n', testtext)
    #Page numbers at the bottom
    testtext = re.sub(r'\nSeite\s\d.*\n', '\n', testtext)
    #Titles
    testtext = re.sub(r'\nSchweizerische Nationalbank.*\n', '\n', testtext)
    testtext = re.sub(r'\nGeldpolitische Lagebeurteilung.*\n', '\n', testtext)
    testtext = re.sub(r'\nJahresende.*\n', '\n', testtext)
    #Tabular Info
    testtext = re.sub(r'\nInflationsprognose.*\%\s*\n', '\n', testtext)
    testtext = re.sub(r'\nInflationsprognose.*\%\s*\n', '\n', testtext)
    testtext = re.sub(r'\nInflationsprognose.*\d{4}\n', '\n', testtext)
    testtext = re.sub(r'\nInflationsprognose.*\d{4}\s*\n', '\n', testtext)
    testtext = re.sub(r'\nDurchschnittliche Jahresteuerung.*\%\n', '\n', testtext)
    testtext = re.sub(r'\n\x0cBeobachtete\sInflation.*\d{4}\s*\n', '\n', testtext)
    testtext = re.sub(r'\nBeobachtete\sInflation.*\d{4}\s*\n', '\n', testtext)
    testtext = re.sub(r'\nInflation\n', '\n', testtext)
    testtext = re.sub(r'\nPrognose.*,\n', '\n', testtext)
    testtext = re.sub(r'\nLibor.*%\n', '\n', testtext)
    #Pure number lines
    testtext = re.sub(r'\n\d+\n', '\n', testtext)
    #Number with whitespace
    testtext = re.sub(r'\n\d+\s*\n', '\n', testtext)
    #Number with comma
    testtext = re.sub(r'\n-*\d+,\d+\n', '\n', testtext)
    #Many numbers with commas
    testtext = re.sub(r'\n(-*\d+,\d+\s*)+\n', '\n', testtext)
    #Many messy numbers with commas and whitespaces
    testtext = re.sub(r'\n[\d\s,-–]\n', '\n', testtext)
    #Quarters
    for i in range(0, 10):
        testtext = re.sub(r'\nQ\d\s*\n', '\n', testtext)
    print (row['date'])
    print(testtext)
    print ('---------')
    return testtext
```

```python
df_lagebeurteilungen['cleantext'] = df_lagebeurteilungen.apply(clean, axis=1)
```

This time, we create a text without any line breaks. Just pure prose from the sentences...

```python
def cleanspaces(row):
    testtext = row['cleantext']
    #Replace all newlines with a whitespace
    testtext = re.sub(r'\n', ' ', testtext)
    #Replace all multi-whitespaces with a single whitespace
    testtext = re.sub(r'\s+', ' ', testtext)
    #Strip it
    testtext = testtext.strip()
    print (row['file'])
    print(testtext)
    print ('---------')
    return testtext    
```

```python
df_lagebeurteilungen['cleantext'] = df_lagebeurteilungen.apply(cleanspaces, axis=1)
```

```python
def savecleanedfinal(row):
    path = 'docs/' + row['file'] + '.final'
    with open(path, 'w') as f:
        f.write(row['cleantext'])
```
