```python
import pandas as pd
%matplotlib inline
```

```python
startyear = 1981
endyear = 2015
```

## Load the data

```python
#Citizenships
df_einb = pd.read_csv('CleanerData/Gem-Einb-Wide-1981-2015.csv')
df_einb.head(2)
```

```python
#Population Levels
df_bev = pd.read_csv('CleanerData/Gem-Bev-Wide-1981-2015.csv')
df_bev.head(2)
```

## Get per capita levels

```python
#First, create a copy of the citizenship dataframe
df_einb_capita = df_einb.copy()
```

```python
df_einb_capita.head(2)
```

```python
#We are doing this by Gemeinde_ID, to make sure nothing is wrong
for ID in df_einb_capita.Gemeinde_ID:
    for year in range(startyear, endyear + 1):
        einb = df_einb[df_einb['Gemeinde_ID'] == ID][str(year)]
        bev = df_bev[df_bev['Gemeinde_ID'] == ID][str(year)]
        einb_capita = einb / bev
        indexnr = df_einb_capita[df_einb_capita['Gemeinde_ID'] == ID].index
        df_einb_capita.loc[indexnr, str(year)] = einb_capita
```

```python
df_einb_capita.head(5)
```

```python
df_einb_capita.to_csv('CleanerData/Gem-EinbCap-Wide-1981-2015.csv')
```
