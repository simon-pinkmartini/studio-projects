### Scrape all the text from the 1. August speeches

```python
#The site where all 1.August speeches are stored
base_url = 'https://www.admin.ch/gov/de/start/dokumentation/reden/ansprachen-zum-nationalfeiertag.html'
response = requests.get(base_url)
doc = BeautifulSoup(response.text, "html.parser")
```

```python
# Select all links pointing to the speeches
linklist_container = doc.find("article", class_="clearfix")
linklist = linklist_container.find_all("li")
```

```python
linklist[0].a['href']
```

```python
linklist[0].text
```

```python
# Store the links in a separate list
article_list = []
for link in linklist:
    current = {}
    current['href'] = "https://www.admin.ch" + link.a['href']
    current['text'] = link.text
    article_list.append(current)
```

```python
# Go over the list and scrape all speech text in raw form
for article in article_list:
    url = article['href']
    response = requests.get(url)
    doc = BeautifulSoup(response.text, "html.parser")
    article['findallp'] = doc.find_all("p")
```

```python
#Some cleaning
for article in article_list:
    print ('-----------')
    for p in article['findallp']:
        print ("***")
        print (p)
```
